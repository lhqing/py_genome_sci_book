```python
import joblib
```

## Serialize any python built-in data structure

Sometimes, you have some special data format that's not in a standard data format (like a table), you can still use joblib to save it into files

```python
spetial_dict = {'apple': 1, 'Location': "CA", 'Data': [1, '2', 5]}

# save it onto disk:
joblib.dump(spetial_dict, 'spetial_dict.lib')
```

```python
# load it back
joblib.load('spetial_dict.lib')
```

## Serialize a whole figure object

We don't usually need to do so, but just explain the idea that almost anything can be saved into file using joblib

```python
import seaborn as sns
import matplotlib.pyplot
```

```python
sns.set(style="ticks")
# Load the example dataset for Anscombe's quartet
df = sns.load_dataset("anscombe")
# Show the results of a linear regression within each dataset
g = sns.lmplot(x="x", y="y", col="dataset", hue="dataset", data=df,
           col_wrap=2, ci=None, palette="muted", height=4,
           scatter_kws={"s": 50, "alpha": 1})
```

```python
# this is a spetial object contain the above plot
g
```

```python
# save it into file
joblib.dump(g, 'sample_figure.lib')
```

```python
# load it back
new_g = joblib.load('sample_figure.lib')
new_g.fig
```
